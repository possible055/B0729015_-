```python
import sympy as sy
import numpy as np
import scipy.stats as st
import scipy as sp
```

```python
#1(a)
π = sy.pi
μ = 0
σ = 1
z = sy.symbols('z')
fz = (1/(sy.sqrt(2*π)*σ))*sy.exp((z-μ)**2/(-2*σ**2))
fz
```

```python
#1(b)
st.norm.cdf(1,μ,σ)-st.norm.cdf(-1,μ,σ)
```

```python
#1(c)
px = (1-0.95)/2
x = st.norm.ppf(0.95+px,μ,σ)
x
```

```python
#1(d)
ν = 1
α = ν/2
β = 2
Γ = sp.special.gamma
q = sy.symbols('q')
fq = (1/((β**α)*Γ(α)))*q**(α-1)*sy.exp(-q/β)
fq
```

```python
#1(e)
μ = α*β
μ
```

```python
#1(f)
σ = sy.sqrt(α*β**2)
σ
```

```python
#1(g)
st.gamma.cdf(x=1, a=α, scale=β)
```
