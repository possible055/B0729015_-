```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from thinkdsp import UncorrelatedUniformNoise

x = UncorrelatedUniformNoise()
x = x.make_wave(duration=1, framerate=10000)
x
```

```python
x.ts,x.ys
x.ys.shape
```

```python
plt.plot(x.ts,x.ys)
```

```python
x.ys.max(), x.ys.mean(),x.ys.std()
```

```python
sx = x.make_spectrum()
sx.fs,sx.hs
sx.hs.shape
```

```python
sx.plot()
```

```python
Sx = sx.make_integrated_spectrum()
Sx.cs.shape
```

```python
Sx.plot_power()
```

```python
Sx.estimate_slope()
```

```python
Sx.plot_power()
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
```

```python
y=x.cumsum()
x.ys.shape, y.ys.shape
```

```python
y.plot()
```

```python
y.ys.max(),y.ys.mean(),y.ys.std()
```

```python
sy=y.make_spectrum()
sy.plot_power()
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
```

```python
sy.make_integrated_spectrum().estimate_slope()
```
